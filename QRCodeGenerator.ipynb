```python
import qrcode as q
from PIL import Image
from IPython.display import display
```

```python
# Data to encode
data = "https://portfolio2-0-f7vh.onrender.com/#contact"

# Generate QR code
qr = q.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)
qr.add_data(data)
qr.make(fit=True)

# Create the image
img = qr.make_image(fill_color="white", back_color="brown")

# Display the image in Jupyter Notebook
display(img)
#img.save(r"C:\Users\HP\Pictures\Screenshots\qrcode.png")
```
